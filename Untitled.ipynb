```python
def compare_by_scatterplot(filename, *exp_evals):
    for curr_exp_evals, label, color in exp_evals:
        fin_vals = [v[-1] for v in curr_exp_evals.values()]
        nfevs = [len(v) for v in curr_exp_evals.values()]
        plt.scatter(nfevs, fin_vals, c=color, label=label)
        
    plt.legend()
    plt.savefig(filename)
    plt.show()
```

```python
example1 = ({1: [0.9, 0.8, 0.7], 2: [0.6 for _ in range(20)]}, 'ex1', 'green')
compare_by_scatterplot('a', example1)
```
